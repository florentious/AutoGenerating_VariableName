```python
# dataset
import os

dir_path = 'data/computers_text/'
dir_list =  [txt for txt in os.listdir(dir_path) if txt.endswith(".txt")]
```

```python
dir_list
```

```python
with open(dir_path+dir_list[0],'r', encoding='utf-16') as f :
    txt = f.read()
```

```python
from bs4 import BeautifulSoup

soupData = BeautifulSoup(txt,'html.parser')
```

```python
soupData
```

```python
body = soupData.findAll('body')
```

```python
body[0].findAll('head')
```

```python
body[0].findAll('p')[0].text
```

```python
def replace(txt) :
    import re
    regEx = re.compile('\(.+?\)')
    result = regEx.sub('',txt)
    
    return result  
    
```

```python
txt_list = []

for i in range(len(body)) :
    for j in range(len(body[i].findAll('p'))) :
        tmp_txt = body[i].findAll('p')[j].text
        txt_list.append(tmp_txt)
```

```python
def list2str(list_) :
    result = ''
    for idx in range(len(list_)) :
        result += list_[idx] + '\n'
        
    return result
```

```python
txt_list
```

```python
with open('convert_.txt','w',encoding='utf-8') as f :
    f.write(list2str(txt_list))
```
