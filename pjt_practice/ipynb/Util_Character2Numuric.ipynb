```python
def char2num(input_,word2idx_path='data/tmp/w2idx.dic') :
    # load_char2idx
    import json
    import numpy as np
    
    with open(word2idx_path,'r') as f :
        tmp = f.read()
    wordDic = json.loads(tmp)
    
#     result = []
    result = np.array([])
    for ch in input_ :
        result = np.append(result,wordDic[ch])
#         result.append(wordDic[ch])
    
    return result.astype('int64')
```

```python
def string2spacingChar(input_) :
    # 빈칸을 '^'로 변환하기
    chars = input_.strip().replace(' ','^')
    
    # SOS : «, EOS : »
    tagged_chars = "«"+chars+"»"
    
#     char_list = ' '.join(list(tagged_chars))
    
    return tagged_chars
```

```python
# 새로운 글자단위 사전 만들기

def newDic(w2i_path='C:/dev/workspace/AutoGenerating_VariableName/pjt_practice/ipynb/data/tmp/w2idx.dic') :
    import json

    with open(w2i_path,'r') as f :
        tmp = f.read()
        wordDic = json.loads(tmp)
        
    ch_list = []
    for _ in range(ord('가'),ord('힣')+1,1) :
        ch_list.append(chr(_))
        
    etc = []
    for ch in wordDic.keys() :
        if not ch in ch_list :
            etc.append(ch)
            
    resDic = {}
    resDic['^'] = 0
    resDic['«'] = 1
    resDic['»'] = 2
    count = 3
    
    for ch in ch_list :
        if not ch in resDic :
            resDic[ch] = count
            count+=1
    for ch in etc :
        if not ch in resDic :
            resDic[ch] = count
            count+=1
            
    jsonString = json.dumps(resDic)
    
    with open('C:/dev/workspace/AutoGenerating_VariableName/pjt_practice/ipynb/data/tmp/newDict.dic','w') as f :
        f.write(jsonString)
        
    return resDic
```
