```python
# convert string to charNum_list
def conv_str2num(input_) :
    def str_pre(input_, input_length=30) :
        # 빈칸을 '^'로 변환하기
        chars = input_.strip().replace(' ','^')
        # SOS : «, EOS : »
        tagged_chars = "«"+chars+"»"
        char_list = ' '.join(list(tagged_chars))
        # 남은칸 PADDING
        for _ in range(input_length - len(tagged_chars)) :
            char_list += ' __PAD__'
        
        return char_list    
        
    def char2num_v2(input_,word2idx_path='data/tmp/newDict.dic') :
        # load_char2idx
        import json
        import numpy as np
        
        with open(word2idx_path,'r') as f :
            wordDic = json.loads(f.read())
        
    #     result = []
        result = np.array([])
        for ch in input_.split(' ') :
            result = np.append(result,wordDic[ch])
    #         result.append(wordDic[ch])
        
        return result.astype('int64')

    return char2num_v2(str_pre(input_))
```

```python
def df2npArr(df) :
    import numpy as np

    res = np.array([])
    for idx in range(len(df)) :
        res = np.append(res, df[idx])
    
    return res
```
